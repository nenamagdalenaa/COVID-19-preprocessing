```python
# Tel het totale aantal documenten per 'dossier_responsible'
responsible_counts = df_dossiers.groupby("dossier_responsible")["document_count"].sum().reset_index()
responsible_counts.columns = ["dossier_responsible", "total_document_count"]
display(responsible_counts)
```

```python
document_type_counts = df_documents["document_sourcetype"].value_counts().reset_index()
document_type_counts.columns = ["document_sourcetype", "count"]
display(document_type_counts)
```

```python
import pandas as pd

# Laad de dataset
df_documents = pd.read_csv("df_documents_final.csv")

# Definieer het basispad
base_path = "/home/nena-meijer/PyCharmMiscProject/unpacked_zipfiles/"

# Maak het volledige pad aan per document
df_documents["document_sourcelocal"] = df_documents.apply(
    lambda row: f"{base_path}dossier_{row['dossier_id']}/{row['document_title']}", axis=1
)

# Sla het aangepaste bestand op
df_documents.to_csv("df_documents_final.csv", index=False)

print("Bijgewerkte dataset opgeslagen als 'df_documents_final.csv'")
```

```python
import pandas as pd
import os

# Laad de dataset
df_documents = pd.read_csv("df_documents_final.csv")

# Controleer of de paden bestaan
df_documents["path_exists"] = df_documents["document_sourcelocal"].apply(os.path.exists)

# Print de eerste paar rijen voor een snelle check
# Print alleen de documenten waarvan het pad niet bestaat
print(df_documents[df_documents["path_exists"] == False][["document_sourcelocal", "path_exists"]])
```
