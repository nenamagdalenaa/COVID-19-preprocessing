```python
import pandas as pd

df_dossiers = pd.read_csv("df_dossiers_final.csv")
df_documents = pd.read_csv("df_ALL_documents.csv")

pd.set_option("display.max_colwidth", None)
display(df_dossiers)
display(df_documents)
```

```python
# Filter df_dossiers op dossiers van het Ministerie van Volksgezondheid, Welzijn en Sport
vws_dossiers = df_dossiers[df_dossiers['dossier_responsible'] == 'ministerie van Volksgezondheid, Welzijn en Sport']
# display(vws_dossiers)

# Selecteer de dossier_id's van deze dossiers
vws_dossier_ids = vws_dossiers['dossier_id']

# Filter df_documents op de dossier_id's die bij VWS horen
vws_documents = df_documents[df_documents['dossier_id'].isin(vws_dossier_ids)]

# display(vws_documents)

print(pd.isnull(vws_documents).sum())

vws_nonna_documents = vws_documents.dropna()
display(vws_nonna_documents)

print(len(vws_nonna_documents.index))

print(pd.isnull(vws_nonna_documents).sum())
```

```python
# Tel het totale aantal documenten per 'dossier_responsible'
responsible_counts = df_dossiers.groupby("dossier_responsible")["document_count"].sum().reset_index()
responsible_counts.columns = ["dossier_responsible", "total_document_count"]
display(responsible_counts)
```

```python
document_type_counts = df_documents["document_sourcetype"].value_counts().reset_index()
document_type_counts.columns = ["document_sourcetype", "count"]
display(document_type_counts)
```

```python
# verkeerd geformatte datums
wrong_formatted_dates = df_documents[df_documents['document_date'].str.startswith('0', na=False)]
print(len(wrong_formatted_dates))
display(wrong_formatted_dates)
```

```python
# missing values
print(pd.isnull(df_documents).sum())
```
