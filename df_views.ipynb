```python
# df_dossministerie van Volksgezondheid, Welzijn en Sportiers.csv
import pandas as pd

df = pd.read_csv("old_csvs/df_dossiers_responsible.csv")
pd.set_option("display.max_colwidth", None)
display(df)
```

```python
# Occurrences van ministeries tellen
responsible_counts = df["dossier_responsible"].value_counts().reset_index()
responsible_counts.columns = ["dossier_responsible", "count"]
display(responsible_counts)
```

```python
# Filteren op ministerie van Volksgezondheid, Welzijn en Sport en totaal aantal documenten berekenen
vws_total_documents = df[df["dossier_responsible"] == "ministerie van Volksgezondheid, Welzijn en Sport"]["document_count"].sum()
print("Document count van minVWS: ", vws_total_documents)
```

```python
import pandas as pd

# CSV-bestanden inlezen
df_dossiers = pd.read_csv("old_csvs/df_dossiers_responsible.csv")
df_documents = pd.read_csv("df_documents.csv")

# Stap 1: Filter de dossiers die horen bij ministerie van Volksgezondheid, Welzijn en Sport
vws_dossier_ids = df_dossiers[df_dossiers["dossier_responsible"] == "ministerie van Volksgezondheid, Welzijn en Sport"]["dossier_id"]
print(len(vws_dossier_ids))

# Stap 2: Filter de documenten die bij deze dossiers horen
vws_documents = df_documents[df_documents["dossier_id"].isin(vws_dossier_ids)]

# Stap 3: Tel het aantal per document_type
document_type_counts = vws_documents["document_type"].value_counts().reset_index()
document_type_counts.columns = ["document_type", "count"]

# Resultaat weergeven
display(document_type_counts)
```
