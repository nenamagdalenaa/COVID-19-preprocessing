```python
import pandas as pd

df_dossiers = pd.read_csv("df_dossiers_final.csv")
df_documents = pd.read_csv("df_ALL_documents.csv")

pd.set_option("display.max_colwidth", None)
display(df_dossiers)
display(df_documents)
```

```python
# Tel het totale aantal documenten per 'dossier_responsible'
responsible_counts = df_dossiers.groupby("dossier_responsible")["document_count"].sum().reset_index()
responsible_counts.columns = ["dossier_responsible", "total_document_count"]
display(responsible_counts)
```

```python
document_type_counts = df_documents["document_sourcetype"].value_counts().reset_index()
document_type_counts.columns = ["document_sourcetype", "count"]
display(document_type_counts)
```

```python
import pandas as pd

# Laad de CSV in Pandas
df = pd.read_csv("df_ALL_documents.csv")

# Sorteer de data op dossier_id en document_id (voor consistentie)
df = df.sort_values(by=["dossier_id", "document_id"])

# Maak een nieuwe doorlopende document_id over alle dossiers
df["document_id"] = range(1, len(df) + 1)

# Sla het bijgewerkte bestand op
df.to_csv("df_documents_sequencial_IDs.csv", index=False)

print("âœ… document_id opnieuw genummerd en opgeslagen als df_documents_sequencial_IDs.csv!")
```
