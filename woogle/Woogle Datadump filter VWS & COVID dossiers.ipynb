```python
# manually set the correct types
bodytext_dtypes = {'dc_publisher_name': str, 'dc_publisher': str, 'foi_documentId': str, 'foi_pageNumber': int,
                  'foi_bodyText': str, 'foi_bodyTextOCR': str, 'foi_hasOCR': bool, 'foi_redacted': float, 'foi_contourArea': float, 'foi_textArea': float, 'foi_charArea': float,
                  'foi_percentageTextAreaRedacted': float, 'foi_percentageCharAreaRedacted': float,
                  'foi_nrWords': int, 'foi_nrChars': int, 'foi_nrWordsOCR ': int, 'foi_nrCharsOCR': int}

dossier_dtypes = {'dc_identifier': str, 'dc_title': str, 'dc_description': str, 'dc_type': str,
                 'foi_type_description': str, 'dc_publisher_name': str, 'dc_publisher': str,
                 'dc_source': str, 'foi_valuation': str, 'foi_requestText': str,
       'foi_decisionText': str, 'foi_isAdjourned': str, 'foi_requester': str}

document_dtypes = {'dc_identifier': str, 'foi_dossierId': str, 'dc_title': str, 'foi_fileName': str,
       'dc_format': str, 'dc_source': str, 'dc_type': str, 'foi_nrPages': int}
```

```python
# Inladen van de datasets met correcte datatypes en indexen
bodytext_df = pd.read_csv("/home/nena-meijer/PyCharmMiscProject/WoogleDumps/woo_bodytext_2i.csv.gz", compression="gzip", dtype=bodytext_dtypes).set_index("foi_documentId")
dossier_df = pd.read_csv("/home/nena-meijer/PyCharmMiscProject/WoogleDumps/woo_dossiers.csv.gz", compression="gzip", parse_dates=[
    "foi_publishedDate", "dc_date_year", "foi_requestDate", "foi_decisionDate", "foi_retrievedDate"
], dtype=dossier_dtypes).set_index("dc_identifier")
document_df = pd.read_csv("/home/nena-meijer/PyCharmMiscProject/WoogleDumps/woo_documents.csv.gz", compression="gzip").set_index("dc_identifier")

# Zet de 'foi_redacted' kolom correct om naar boolean (vanwege NaN-waarden eerst naar float)
bodytext_df["foi_redacted"] = bodytext_df["foi_redacted"].fillna(0).astype(bool)
```

```python
# Basisinfo & kolomnamen tonen
print("Bodytext dataset info:")
print(bodytext_df.info(), "\n")

print("Dossiers dataset info:")
print(dossier_df.info(), "\n")

print("Documenten dataset info:")
print(document_df.info(), "\n")
```

## MinVWS dossiers eruit filteren

```python
# Unieke namen van ministeries bekijken om te zien hoe VWS wordt aangeduid
print(dossier_df["dc_publisher_name"].unique())
```

```python
import pandas as pd

# Zorg ervoor dat dossier_df en document_df correct zijn ingeladen
# Filter alle dossiers van VWS uit dossier_df
vws_dossiers = dossier_df[dossier_df["dc_publisher_name"].str.contains(
    "Ministerie Van Volksgezondheid, Welzijn En Sport", case=False, na=False
)]

# Show de eerste 10 dossiers
vws_dossiers.head(10)
```

## COVID gerelateerde dossiers uit MinVWS dossiers filteren

```python
import pandas as pd
from IPython.core.display import display, HTML

# Definieer zoekwoorden voor filtering
zoekwoorden = ["covid", 
               "corona", 
               "sars", 
               "avondklok", 
               "social distancing", 
               "mondkapje", 
               "OpenVWS",
                "PCR",
              "quarantaine",
              "zelftest",
              "reisbeleid",
              "coronabeleid",
              "Sywert",
              "Noodmaatregel Overbrugging voor Werkgelegenheid",
              "NOW",
              "pandemic",
              "pandemie",
              "OMT-V",
               "Besluit op Woo-verzoek over audiobestanden vergaderingen OMT",
               "bivalente vaccins",
               "coronavaccinatie",
               "coronavaccins"
               "booster",
               "Bezetting en predicties"
               "AstraZeneca",
               "lockdown",
               "moderna",
               "pfizer",
               "mRNA",
               "teststraat",
               "IC-opnames",
               "antigeentest",
               "rioolwater",
               "desinformatie",
               "gesloten scholen",
               "Viruswaarheid",
               "viruswaanzin",
               "Kern over de periode april 2020"
              ]
# Maak een regex-patroon van de zoekwoorden
regex_pattern = "|".join(zoekwoorden)

# Filter de dossiers waarvan de titel of beschrijving een van de zoekwoorden bevat
gefilterde_dossiers = vws_dossiers[
    vws_dossiers["dc_title"].str.contains(regex_pattern, case=False, na=False) |
    vws_dossiers["dc_description"].str.contains(regex_pattern, case=False, na=False)
]

# Controleer het aantal gevonden dossiers
print(f"Aantal dossiers na filtering: {len(gefilterde_dossiers)}")

# Maak de Pandas DataFrame scrollbaar
styled_df = gefilterde_dossiers.style.set_table_attributes(
    'style="width:100%; display:block; overflow-x: auto; max-height: 500px; overflow-y: scroll;"'
)

# Zet de tabel zo breed mogelijk in Jupyter Notebook
display(HTML("<style>table { width:100% !important; }</style>"))

# Toon de DataFrame met styling
display(styled_df)
```

```python
# Tellen van het aantal dossiers per type (Woo-informatie categorie)
type_tellingen = gefilterde_dossiers["dc_type_description"].value_counts()

# Resultaat weergeven
print(type_tellingen)

# Haal alle waarden uit de kolom 'dc_title' op en zet ze in een lijst
woogle_vws_dossiers_titels = gefilterde_dossiers["dc_title"].dropna().tolist()
```

```python
# documenten koppelen aan dossiers

covid_documenten = document_df[
    document_df["foi_dossierId"].isin(gefilterde_dossiers.index)
]

# Maak de Pandas DataFrame scrollbaar
styled_df = covid_documenten.style.set_table_attributes(
    'style="width:100%; display:block; overflow-x: auto; max-height: 500px; overflow-y: scroll;"'
)

# Zet de tabel zo breed mogelijk in Jupyter Notebook
display(HTML("<style>table { width:90% !important; }</style>"))

display(styled_df)

print("Aantal documenten: ", len(covid_documenten))
```
