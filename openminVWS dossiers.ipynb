```python
import os
import zipfile

dossiers_folder = "/home/nena-meijer/PyCharmMiscProject/VWS_dossiers_zip"

zip_count = 0
total_documents = 0
zip_names = []  # Lijst voor ZIP-bestandsnamen zonder extensie

# Controleren of de map bestaat
if os.path.exists(dossiers_folder):
    # Lijst van bestanden in de map
    files = os.listdir(dossiers_folder)
    
    for file in files:
        if file.lower().endswith(".zip"):  # Case-insensitive check
            zip_count += 1
            zip_name = os.path.splitext(file)[0]  # Verwijdert de .zip extensie
            zip_names.append(zip_name)
            
            zip_path = os.path.join(dossiers_folder, file)
            
            try:
                # Open de ZIP en tel de bestanden (exclusief metadata.json)
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    documenten = [f for f in zip_ref.namelist() if not f.lower().endswith("metadata.json")]
                    total_documents += len(documenten)
            except zipfile.BadZipFile:
                print(f"Waarschuwing: {file} is geen geldig ZIP-bestand en wordt overgeslagen.")

print(f"Aantal dossiers: {zip_count}")
print(f"Totaal aantal documenten (exclusief metadata.json): {total_documents}")
```

```python
import zipfile
import os
import pathlib

# Pad naar de ZIP-file (pas dit aan als nodig)
zip_path = "/home/nena-meijer/PyCharmMiscProject/VWS_dossiers_zip/Woo-besluit inzake openbaarmaking chat- en smsberichten.zip" # Zorg ervoor dat 'first_dossier' het juiste pad bevat

# Uitpakmap
extract_folder = "unzipped_files"
os.makedirs(extract_folder, exist_ok=True)

# Uitpakken van de ZIP-file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

# Lijst van bestanden in de uitgepakte map
files_list = list(pathlib.Path(extract_folder).rglob("*.*"))  # Alle bestanden in de map

# Controleer per bestand of er metadata is
from pathlib import Path
from mutagen import File  # Voor media metadata
import PyPDF2  # Voor PDF metadata
from PIL import Image  # Voor afbeeldingen

metadata_info = {}

for file_path in files_list:
    file_path = str(file_path)  # Zorg voor een stringformaat

    metadata = None
    file_extension = Path(file_path).suffix.lower()

    try:
        if file_extension in [".mp3", ".flac", ".wav", ".ogg"]:  # Media bestanden
            audio = File(file_path)
            metadata = audio.tags if audio else None
        
        elif file_extension == ".pdf":  # PDF bestanden
            with open(file_path, "rb") as f:
                pdf = PyPDF2.PdfReader(f)
                metadata = pdf.metadata

        elif file_extension in [".jpg", ".jpeg", ".png", ".tiff", ".bmp"]:  # Afbeeldingen
            with Image.open(file_path) as img:
                metadata = img.info

        elif file_extension in [".json", ".xml", ".txt"]:  # Probeer tekst-gebaseerde metadata te lezen
            with open(file_path, "r", encoding="utf-8") as f:
                metadata = f.read()[:500]  # Eerste 500 tekens om metadata te inspecteren
        
        if metadata:
            metadata_info[file_path] = metadata

    except Exception as e:
        metadata_info[file_path] = f"Fout bij lezen metadata: {e}"

# Print de resultaten
for doc, meta in metadata_info.items():
    print(f"Document: {doc}\nMetadata: {meta}\n{'-'*50}")
```
